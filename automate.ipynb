```python
import requests
import os

# Define base URL go to the aemo site
root_url = 'https://aemo.com.au/aemo/data/nem/priceanddemand/PRICE_AND_DEMAND_'

# create a directory where you want to keep the files 
folder = "directory to your folder"

#create the year tag list with format YYYYMM

years = [['202401','202402','202403','202404','202405','202406','202407','202408','202409','202410','202411','202412'],
         ['202301','202302','202303','202304','202305','202306','202307','202308','202309','202310','202311','202312'],
         ['202201','202202','202203','202204','202205','202206','202207','202208','202209','202210','202211','202212']]

#create the list comprehension to loop trough the years
flat = [value for sublist in years for value in sublist]

#list of available regions
region =['NSW1','QLD1', 'VIC1','SA1','TAS1']    

## Use headers to simulate a browser request (some sites block non-browser requests)
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/137.0.0.0 Safari/537.36',
    'Referer': 'https://aemo.com.au/aemo/apps/visualisation/index.html',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
    'Connection': 'keep-alive',
    # 'Cookie': 'your-cookie-here'
 } 

#created function to fetch the string based on year and month 
def fileloader(flat,region):
    for i in flat:
        for k in region:
            

    # Construct the full file URL 
            url = f"{root_url}{i}_{k}.csv"
    
            file_name = f"PRICE_AND_DEMAND_{i}_{k}.csv"
            # in this version we use and outside path but the files shoud be in a subfolder for the next step of the project
            output_path = os.path.join(folder, file_name)
            
            
            # Make sure the output directory exists
            os.makedirs("ETL2", exist_ok=True)
        
            response = requests.get(url, headers=headers)
            if response.status_code == 200:
                with open(output_path, 'wb') as f:
                    f.write(response.content)
                print(f"Downloaded: {file_name}")
            else:
                print(f"Failed: {file_name} (status code: {response.status_code})")
#call the function                
fileloader(flat, region)
```

```python
import pandas as pd
import glob
import os
import re


#create the path to your folder with the downloaded csv files 
all_files = glob.glob(os.path.join("your folder path", "PRICE_AND_DEMAND_*.csv"))

# Loop through and read each file groupe files by year using regex
files_2022 = [f for f in all_files if re.search(r'PRICE_AND_DEMAND_2022\d{2}_', os.path.basename(f))]
files_2023 = [f for f in all_files if re.search(r'PRICE_AND_DEMAND_2023\d{2}_', os.path.basename(f))]
files_2024 = [f for f in all_files if re.search(r'PRICE_AND_DEMAND_2024\d{2}_', os.path.basename(f))]

#create a Function to concatenate list of files
def concat_files(file_list):
    return pd.concat([pd.read_csv(f) for f in file_list], ignore_index=True)

# Create separate DataFrames
df_2022 = concat_files(files_2022)
df_2023 = concat_files(files_2023)
df_2024 = concat_files(files_2024)


#Save to new CSVs
df_2022.to_csv(os.path.join("your folder path", "PRICE_AND_DEMAND_2022_ALL.csv"), index=False)
df_2023.to_csv(os.path.join("your folder path", "PRICE_AND_DEMAND_2023_ALL.csv"), index=False)
df_2024.to_csv(os.path.join("your folder path", "PRICE_AND_DEMAND_2024_ALL.csv"), index=False)
```
